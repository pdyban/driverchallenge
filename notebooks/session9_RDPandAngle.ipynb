```python
pathToDriverData = '../../drivers_npy/'
pathToCurveData = '../../curve_npy/'

import DriverChallengeHelperFunctions as helpers
import DriverDataIO as io
import DriverChallengeVisualization as vis
import sys

sys.path.append("../../RDP/")

rdp = __import__("rdp")

import numpy
import os
scigor = __import__("Session 4 scigor side curvature speed feature")

%load_ext autoreload
%autoreload 2
```

```python
rdpFactor = 2
```

```python
def getFeatureValForDriver(driverid):
    feature_array = []
    for i in range(1,201):
        _trip = io.get_trip_npy(driverid,i, pathToDriverData)
        
        #rdp
        _rdpTrip = rdp.rdp(_trip, rdpFactor)
        
        _angle_feature_elements = scigor.angle_feature(_rdpTrip)
        
        if(len(_angle_feature_elements) > 0):
            _drive_mean = numpy.mean(_angle_feature_elements)
            feature_array.append((i, _drive_mean))
    features = numpy.array(feature_array)
    return features
```

```python
for driver in io.list_all_drivers(pathToDriverData):
    featureVal = getFeatureValForDriver(driver)
    print str(driver) + " " + str(featureVal)
```
